```python
# Importação das bibliotecas\n",
import pandas as pd
import seaborn as srn
#pip install seaborn (executar no Anaconda Prompt)\n",
import matplotlib.pyplot as plt
```

```python
# Carregamento de outra base de dados\n",
base2 = pd.read_csv('chicken.csv')
base2.head()
```

```python
# Criação de novo dataframe agrupando o atributo 'feed'\n",
agrupado = base2.groupby(['feed'])['weight'].sum()
agrupado
```

```python
# Novo dataframe somente para testar os filtros do pandas\n",
teste = base2.loc[base2['feed'] == 'horsebean']
teste
```

```python
# Histrograma considerando somente o valor 'horsebean'\n",
#apenas o grafico de densidade neste caso\n",
srn.kdeplot(base2.loc[base2['feed'] == 'horsebean'].weight).set(title='horsebean')
```

```python
# Histrograma considerando somente o valor 'casein'\n",
srn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True).set(title='casein')
```

```python
#impresssão em gráfico 2 x 3 - 3 linhas 2 colunas\n",
plt.figure()
plt.subplot(3,2,1)
srn.histplot(base2.loc[base2['feed'] == 'horsebean'].weight, kde=True).set(title='horsebean')
plt.subplot(3,2,2)
srn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True).set(title='casein')
plt.subplot(3,2,3)
srn.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde=True).set(title='linseed')
plt.subplot(3,2,4)
srn.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde=True).set(title='meatmeal')
plt.subplot(3,2,5)
srn.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde=True).set(title='soybean')
plt.subplot(3,2,6)
srn.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde=True).set(title='sunflower')
#ajusta o layout para não haver sobreposição\n",
plt.tight_layout()
```
